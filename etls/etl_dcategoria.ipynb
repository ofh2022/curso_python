```python
from db_connection import get_postgres_connection, get_sqlserver_connection
import pandas as pd
```

```python
#extraindo dados do banco de produ√ß√£o


def extract_categoria(limit_days: int = 7):

    query = f"""
        SELECT id,
        descricao
FROM vendas.categoria;
    """

    conn = get_postgres_connection()
    df = pd.read_sql(query, conn)
    conn.close()

    return df
```

```python
df = extract_categoria()

print(df.dtypes)
```

```python
# =====================
# 3. LOAD (SQL Server)
# =====================
mssql_conn = get_sqlserver_connection()
cursor = mssql_conn.cursor()

insert_sql = """
    INSERT INTO dw.categoria (
        id_categoria,
        descricao
    )
    VALUES (?, ?)
"""

for _, row in df.iterrows():
    cursor.execute(insert_sql, (
    row["id"],
    row["descricao"]
    ))

mssql_conn.commit()
cursor.close()
mssql_conn.close()

print("‚úî Dados carregados no SQL Server")
print("üèÅ ETL finalizado com sucesso")
```
